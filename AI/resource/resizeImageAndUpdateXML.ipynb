```python
from PIL import Image
import os
import xml.etree.ElementTree as ET
```

```python
## 3개짜리 구분 여부
def include_three(file):
    name = file.split('_')
    if(name[2]=='s'):
        return False
    else:
        return True
```

```python
# 경로 저장
DIR_PATHS = './raw'

#IMAGE_PATHS = sorted(list(DIR_PATHS.glob("*.jpg")))
#XML_PATHS = sorted(list(DIR_PATHS.glob("*.xml")))

xml_list = []
img_list = []
file_list = os.listdir(DIR_PATHS)
isInclude = False # 3개짜리 사진도 포함할거면 True로 변경

for file in file_list:
    if '.xml' in file:
        if(isInclude == False):
            if(include_three(file) == False):
                xml_list.append(file)
        else:
            xml_list.append(file)
    elif '.jpg' in file:
        if(isInclude == False):
            if(include_three(file) == False):
                img_list.append(file)
        else:
            img_list.append(file)
```

```python
# 파일 경로 및 개수 확인
print(xml_list)
print(len(xml_list))
```

```python
## 이미지 사이즈 수정 : 2988 / 320 = 9.3375
num = 1
for img_file in img_list:
    img = Image.open(DIR_PATHS+"\\"+img_file)
    img_resize = img.resize((int(img.width / 9.3375), int(img.height / 9.3375)))
    img_resize.save("RESIZE/" + img_file)
    
    print("[" + str(num) + "]" + img_file + "[success]")
    num += 1
    
print("============ DONE = total : "+str(num)+" ============")
```

```python
## TAG 수정 : size 320으로 resize & bnd 좌표 수정 (int형)
num = 1
for xml_file in xml_list:
    targetXML = open(DIR_PATHS+ "/" + xml_file, 'rt', encoding='UTF8')

    tree = ET.parse(targetXML)

    root = tree.getroot()
    wh_tag = root.findall('size')[0]
    bnd_tag = root.findall("object")[0]
    name_tag = bnd_tag.findall("name")[0]
    bnd_tag = bnd_tag.findall("bndbox")[0]     

    ##수정할 부분
    
    name = name_tag.text
    if name == "롯데밀키스250ML":
        name = "milkis"
    elif name == "롯데칠성사이다190ML":
        name = "cider"
        
    name_tag.text = name
    
    w_tag = wh_tag.find('width') 
    h_tag = wh_tag.find("height")
    xmin_tag = bnd_tag.find("xmin")
    ymin_tag = bnd_tag.find("ymin")
    xmax_tag = bnd_tag.find("xmax")
    ymax_tag = bnd_tag.find("ymax")
    
    w_tag.text = '320'
    h_tag.text = '320'
    xmin_tag.text = str(int(float(xmin_tag.text)/9.3375))
    ymin_tag.text = str(int(float(ymin_tag.text)/9.3375))
    xmax_tag.text = str(int(float(xmax_tag.text)/9.3375))
    ymax_tag.text = str(int(float(ymax_tag.text)/9.3375))
    
    print("[" + str(num) + "]" + xml_file + "[success]")
    
    tree.write("RESIZE/"+xml_file, encoding = 'utf-8')
    num += 1

print("============ DONE = total : "+str(num)+" ============")
```
