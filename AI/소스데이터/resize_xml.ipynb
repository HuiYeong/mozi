```python
import os
import xml.etree.ElementTree as ET
```

```python
targetDir = r"C:\Users\multicampus\Desktop\싸피\2차프로젝트\dataset\resize_label"
target2Dir = r"C:\Users\multicampus\Desktop\싸피\2차프로젝트\dataset\resize_label2"
num = 1

##targetDir에서 .xml파일 이름들 리스트로 가져오기
file_list = os.listdir(targetDir)
xml_list = []
for file in file_list:
    if '.xml' in file:
        xml_list.append(file)
```

```python
print(xml_list)
print(len(xml_list))
```

```python
## size 80으로 resize & round 2
for xml_file in xml_list:
    target_path = targetDir + "\\" + xml_file
    targetXML = open(targetDir+ "\\" + xml_file, 'rt', encoding='UTF8')

    tree = ET.parse(targetXML)

    root = tree.getroot()
    wh_tag = root.findall('size')[0]
    bnd_tag = root.findall("object")[0]
    bnd_tag = bnd_tag.findall("bndbox")[0]
    '''for child in root.iter():
        print (child.tag)'''
        

    ##수정할 부분
    w_tag = wh_tag.find('width') 
    h_tag = wh_tag.find("height")
    xmin_tag = bnd_tag.find("xmin")
    ymin_tag = bnd_tag.find("ymin")
    xmax_tag = bnd_tag.find("xmax")
    ymax_tag = bnd_tag.find("ymax")
    
    w_tag.text = '80'
    h_tag.text = '80'
    xmin_tag.text = str(round((float(xmin_tag.text)/37.35), 2))
    ymin_tag.text = str(round((float(ymin_tag.text)/37.35), 2))
    xmax_tag.text = str(round((float(xmax_tag.text)/37.35), 2))
    ymax_tag.text = str(round((float(ymax_tag.text)/37.35), 2))
    
    print("[" + str(num) + "]" + xml_file + "[success]")
    
    tree.write(target_path, encoding = 'utf-8')
    num += 1

print("finished")
```
